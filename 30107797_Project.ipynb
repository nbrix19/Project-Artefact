**Year-on-Year Percentage Change in Cholera Cases**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Ensure all cholera case columns are numeric
for year in years:
    column_name = f'CholCase{str(year)[-2:]}'  
    cholera_cases.loc[:, column_name] = pd.to_numeric(cholera_cases[column_name], errors='coerce')
    

# Sum cholera cases across all states for each year
total_cholera_cases = cholera_cases[years].sum()

# Calculate year-on-year percentage change
pct_change = total_cholera_cases.pct_change() * 100

# Increase the figure size for better visibility
plt.figure(figsize=(8, 6))

# Define a color gradient for the bars based on total cases
colors = plt.cm.coolwarm(np.linspace(0.2, 0.8, len(total_cholera_cases)))

# Create the bar plot
bars = plt.bar(range(2014, 2024), total_cholera_cases, color=colors, alpha=0.85, edgecolor='black', width=0.6)

# Loop through bars and annotate them with total cases and percentage change
for i, bar in enumerate(bars):
    height = bar.get_height()

    # Annotate total cholera cases inside the bars, placing based on height for readability
    if height > 5000:
        plt.text(bar.get_x() + bar.get_width() / 2, height - 4000, f'{int(height):,}',
                 ha='center', va='top', fontsize=10, color='white', fontweight='bold')
    else:
        plt.text(bar.get_x() + bar.get_width() / 2, height + 500, f'{int(height):,}',
                 ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')

    # Annotate percentage change above the bars
    if i > 0:  # Skip the first bar (no % change for 2014)
        prev_height = total_cholera_cases[i-1]
        if prev_height == 0:  # Handle division by zero
            plt.text(bar.get_x() + bar.get_width() / 2, height + 2000,
                     'New cases', ha='center', va='bottom', fontsize=10,
                     fontweight='bold', color='purple')
        else:
            change = pct_change[i]
            if not pd.isna(change):  # Only annotate valid percentage changes
                # Use color coding: green for increase, red for decrease
                color = 'green' if change > 0 else 'red'
                plt.text(bar.get_x() + bar.get_width() / 2, height + 2000,
                         f'{change:+.2f}%', ha='center', va='bottom', fontsize=10,
                         fontweight='bold', color=color)

# Customize plot appearance
plt.title("Total Cholera Cases and Year-on-Year % Change (2014-2023)", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Cholera Cases", fontsize=12)

# Set x-axis ticks and labels
plt.xticks(ticks=range(2014, 2024), labels=range(2014, 2024))

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Set y-axis limits to create space for annotations
plt.ylim(0, total_cholera_cases.max() + 15000)  # Add some space above the bars for annotations

# Apply tight layout to ensure proper spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Reshape the data to long format
popden_long = data.melt(value_vars=['PopDen14', 'PopDen15', 'PopDen16', 'PopDen17', 'PopDen18', 'PopDen19', 'PopDen20', 'PopDen21', 'PopDen22', 'PopDen23'],
                        var_name='Year', value_name='PopDen')
cholcase_long = data.melt(value_vars=['CholCase14', 'CholCase15', 'CholCase16', 'CholCase17', 'CholCase18', 'CholCase19', 'CholCase20', 'CholCase21', 'CholCase22', 'CholCase23'],
                          var_name='Year', value_name='CholCase')

# Combine the reshaped data
data_long = pd.concat([popden_long['PopDen'], cholcase_long['CholCase']], axis=1)

# Define the dependent and independent variables
X = data_long['PopDen']
y = data_long['CholCase']

# Add a constant to the independent variables
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Get the summary of the model
print(model.summary())

# Create a scatter plot
plt.scatter(data_long['PopDen'], data_long['CholCase'], color='blue', label='Data points')

# Fit a linear regression line
coefficients = np.polyfit(data_long['PopDen'], data_long['CholCase'], 1)
poly = np.poly1d(coefficients)
plt.plot(data_long['PopDen'], poly(data_long['PopDen']), color='red', label='Fit line')
```
